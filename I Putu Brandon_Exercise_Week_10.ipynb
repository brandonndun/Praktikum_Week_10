# Association Rule for Store Dataset

In this case study, we will explore how association rule can be used to analyze the items that are usualy purcased together.

## Load Data

We will use the dataset of the transaction in a certain store. You can get the dataset here: 
https://gist.githubusercontent.com/Harsh-Git-Hub/2979ec48043928ad9033d8469928e751/raw/72de943e040b8bd0d087624b154d41b2ba9d9b60/retail_dataset.csv

```python
# load the data set ans show the first five transaction
dataset = pd.read_csv("https://gist.githubusercontent.com/Harsh-Git-Hub/2979ec48043928ad9033d8469928e751/raw/72de943e040b8bd0d087624b154d41b2ba9d9b60/retail_dataset.csv")
dataset.head()
```

# Get the set of product that has been purchased


```python
set_items = pd.unique(dataset.values.ravel())
setunique = set(set_items)
print(set_items)
```

## Preprocess Data

In this step, we will transform our dataset so that we will have a one hot encoding based on the purchased products.

```python
#create an itemset based on the products
itemset = set(set_items)
# encoding the feature
encoded_vals = []
for index, row in dataset.iterrows():
    rowset = set(row) 
    labels = {}
    uncommons = list(itemset - rowset)
    commons = list(itemset.intersection(rowset))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
```

```python
  # create new dataframe from the encoded features
dataset1 = pd.DataFrame(encoded_vals)
  # show the new dataframe
dataset1.head()
```

Since, the encoded dataframe consist of the empty column. We will drop the NaN column or select all columns other than the first column.

```python
dataset1.drop(columns=dataset1.columns[0], axis=1, inplace=True)
```

## Apriori Algorithm

We will use appriori algorithm to determine the frequently purchased products. 
For this case study, we will min_support=0.2

```python
from mlxtend.frequent_patterns import apriori, association_rules
frq_items = apriori(dataset1, min_support = 0.2, use_colnames = True)
frq_items
```

The we will generate association rule of the frequent itemset based on confidence level with the threshold=0.6

```python
rules = association_rules(frq_items, metric ="confidence", min_threshold = 0.6)
rules
```
